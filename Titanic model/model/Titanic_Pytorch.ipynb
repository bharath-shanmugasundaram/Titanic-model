```python
import torch as t
import torch.nn as nn
import numpy as np
import pandas as pd

df = pd.read_csv("/Users/bhara-zstch1566/Deep Learning/Projects/AUG-WEEK-2/P2/dataset/titanic/test.csv")
Y = pd.read_csv("/Users/bhara-zstch1566/Deep Learning/Projects/AUG-WEEK-2/P2/dataset/titanic/gender_submission.csv")
Y = Y.drop("PassengerId", axis=1).to_numpy()

df = df.drop(["PassengerId", "Name"], axis=1)

df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
df["Embarked"] = df["Embarked"].map({'C': 1, 'Q': 2, 'S': 3})
df = df.fillna(0)

def ticket_to_num(ticket):
    parts = ticket.split()
    return int(parts[-1]) if parts[-1].isdigit() else 0
df["Ticket"] = df["Ticket"].apply(ticket_to_num)

df["Cabin"] = df["Cabin"].apply(lambda x: 0 if x == 0 else 1)

X = df.to_numpy()
X = np.nan_to_num(X).astype(float)

print("Feature matrix shape:", X.shape)
print("Label shape:", Y.shape)








```

```python
X = (X - X.mean(axis=0)) / (X.std(axis=0) + 1e-8)

X = t.from_numpy(X).type(t.float32)
Y = t.tensor(Y).type(t.float32)

print(X.shape, Y.shape)
```

```python
class LogisticRegression(nn.Module):
    def __init__(self, input_features):
        super(LogisticRegression, self).__init__()
        self.linear = nn.Linear(input_features, 1)

    def forward(self, x):
        return self.linear(x)  

model = LogisticRegression(X.shape[1])
```

```python

criterion = nn.BCEWithLogitsLoss()  
optimizer = t.optim.SGD(model.parameters(), lr=0.01)
```

```python
num_epochs = 1000
for epoch in range(num_epochs):
    y_pred = model(X)
    loss = criterion(y_pred.view(-1), Y.view(-1))   # flatten
    
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    
    if (epoch+1) % 100 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
with t.no_grad():
    y_predicted = model(X)
    y_predicted_cls = t.sigmoid(y_predicted).round()
    acc = y_predicted_cls.eq(Y).sum() / float(Y.shape[0])
    print(f'Final Accuracy = {acc:.4f}')
```
